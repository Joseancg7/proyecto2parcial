```python
#Librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

#Constantes cinéticas
umax = 0.4674
Ki = 0.004
Ks = 4.1
Xmax = 47.3
m = 1
Pmax = 75
n = 1.5
Yx = 0.054
Ypx =10.07
mx = 0.2
mp = 0.1
#Ecuaciones cinéticas
def cinetica(y,t,umax,Ki,Ks,Xmax,m,Pmax,n,Yx,Ypx,mx,mp):
    X = y[0]
    S = y[1]
    P = y[2]
    u = np.exp(-Ki*S)*(umax*S)/(Ks+S)*(1-X/Xmax)**m*(1-P/Pmax)**n
    dX = X*u
    dS = -dX/Yx - mx*X
    dP = Ypx*dX + mp*X
    return np.array([dX,dS,dP])

# Condiciones iniciales
y0 = np.array([0.7,131.221,8.424])

# Tiempo en el que se desarrolla la cinética
t = np.linspace(0, 25,10000)

# Solución numérica
y = odeint(cinetica, y0, t, args=(umax,Ki,Ks,Xmax,m,Pmax,n,Yx,Ypx,mx,mp))

# Extrar e y s de la solucion
X = y[:,0]
S = y[:,1]
P = y[:,2]


fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('tiempo (h)')
ax1.set_ylabel('Concentración Sustrato', color=color)
plt.plot(t, S,label='Concentración Sustrato')
plt.axis(ymax = 140)
ax1.tick_params(axis='y', labelcolor=color)
plt.legend(loc='upper left')

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Concentración Producto y Biomasa', color=color)  # we already handled the x-label with ax1
plt.plot(t, X, label='Concentración Biomasa')
plt.plot(t, P, label='Concentración Producto')
plt.axis(ymax = 90)
plt.legend(loc='upper right')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.show()
```
